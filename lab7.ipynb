```python
time_points = np.linspace(0, 1, 500, endpoint=False)
original_signal = np.sin(2 * np.pi * 3 * time_points)

noise_level = 0.4
noise = noise_level * np.random.randn(len(time_points))
noisy_signal = original_signal + noise

averaged_signal = np.zeros_like(original_signal)
window_size = 20
for i in range(window_size, len(time_points) - window_size):
    averaged_signal[i] = (2 * window_size + 1) ** -1 * np.sum(noisy_signal[i - window_size:i + window_size + 1])

plt.figure(figsize=(10, 6))

plt.subplot(2, 3, 1)
plt.plot(time_points, original_signal)
plt.xlabel('Time')
plt.ylabel('Signal Value')
plt.title('Original Signal')

plt.subplot(2, 3, 2)
plt.plot(time_points, noisy_signal)
plt.xlabel('Time')
plt.ylabel('Noisy Signal')
plt.title('Signal with Noise')

plt.subplot(2, 3, 3)
plt.plot(time_points, averaged_signal)
plt.xlabel('Time')
plt.ylabel('Averaged Signal')
plt.title('Averaged Signal')

spectrum_original = np.abs(np.fft.fft(original_signal))
freq_original = np.fft.fftfreq(len(original_signal))
plt.subplot(2, 3, 4)
plt.plot(freq_original, spectrum_original)
plt.title('Spectrum of Original Signal')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.grid()

spectrum_noisy = np.abs(np.fft.fft(noisy_signal))
freq_noisy = np.fft.fftfreq(len(noisy_signal))
plt.subplot(2, 3, 5)
plt.plot(freq_noisy, spectrum_noisy)
plt.title('Spectrum of Noisy Signal')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.grid()

spectrum_averaged = np.abs(np.fft.fft(averaged_signal))
freq_averaged = np.fft.fftfreq(len(averaged_signal))
plt.subplot(2, 3, 6)
plt.plot(freq_averaged, spectrum_averaged)
plt.title('Spectrum of Averaged Signal')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
signal = np.random.rand(100)

threshold = 0.5
filtered_signal = np.where(signal<threshold, signal, 0)

window_size = 5
median_filtered_signal = np.copy(filtered_signal)

for i in range(window_size, len(filtered_signal) - window_size):
    median_filtered_signal[i] = np.median(filtered_signal[i - window_size:i + window_size + 1])

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(signal, label='Original Signal')
plt.xlabel('Time')
plt.ylabel('Signal Value')
plt.title('Original Signal')

plt.subplot(1, 3, 2)
plt.plot(filtered_signal, label='Signal after Threshold Filtering')
plt.xlabel('Time')
plt.ylabel('Signal Value')
plt.title('Signal after Threshold Filtering')

plt.subplot(1, 3, 3)
plt.plot(median_filtered_signal, label='Median Filter')
plt.xlabel('Time')
plt.ylabel('Signal Value')
plt.title('Median Filter with Threshold Filtering')

plt.tight_layout()

plt.show()
```

```python
from scipy.fft import fft, ifft

fs = 1000
t = np.arange(0, 10, 1/fs)
signal = np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 10 * t)
original_signal = signal.copy()
signal[5*fs:7*fs] = 0
window_size = int(fs * 1)

spectrum_before = fft(original_signal[:5*fs])
spectrum_after = fft(signal[7*fs:])
min_len = min(len(spectrum_before), len(spectrum_after))
average_spectrum = 0.5 * (spectrum_before[:min_len] + spectrum_after[:min_len])

reconstructed_signal = np.concatenate([
    np.real(ifft(spectrum_before)),
    np.real(ifft(average_spectrum)),
    np.real(ifft(spectrum_after))
])

plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(t, original_signal)
plt.title('Original Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, signal)
plt.title('Original Signal with Gap')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, reconstructed_signal[:len(t)])
plt.title('Reconstructed Signal using Spectral Interpolation')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```

Task 6

```python
from scipy.signal import butter, filtfilt

def downsample_signal(signal, N):
    fs_old = 1000
    fs_new = fs_old / N

    nyquist = 0.5 * fs_old
    low_cutoff = 0.5 * fs_new
    b, a = butter(4, low_cutoff/nyquist, btype='low')
    filtered_signal = filtfilt(b, a, signal)

    downsampled_signal = filtered_signal[::N]

    return downsampled_signal

fs = 1000
t = np.arange(0, 1, 1/fs)
original_signal = np.cos(2 * np.pi * 50 * t) + 0.5 * np.cos(2 * np.pi * 100 * t)

N = 4
downsampled_signal = downsample_signal(original_signal, N)

plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, original_signal)
plt.title('Original Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t[::N], downsampled_signal)
plt.title(f'Downsampled Signal (N={N})')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
t1 = np.arange(0, 1, 0.035)
t2 = np.arange(0, 1, 0.05)
t3 = np.arange(0, 1, 0.08)

signal1 = np.cos(2 * np.pi * 35 * t1)
signal2 = np.cos(2 * np.pi * 50 * t2)
signal3 = np.cos(2 * np.pi * 80 * t3)

min_fs = max(1/0.035, 1/0.05, 1/0.08)
t_resampled = np.linspace(0, 1, int(1 * min_fs))
cosine1 = np.interp(t_resampled, t1, signal1)
cosine2 = np.interp(t_resampled, t2, signal2)
cosine3 = np.interp(t_resampled, t3, signal3)

plt.figure(figsize=(12, 8))

plt.subplot(231)
plt.plot(t1, signal1)
plt.title("Sampling Frequency 35 Hz")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid()

plt.subplot(232)
plt.plot(t2, signal2)
plt.title("Sampling Frequency 50 Hz")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid()

plt.subplot(233)
plt.plot(t3, signal3)
plt.title("Sampling Frequency 80 Hz")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid()

plt.subplot(234)
plt.plot(t_resampled, cosine1)
plt.title("Sampling Frequency 35 Hz (Interpolated)")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid()

plt.subplot(235)
plt.plot(t_resampled, cosine2)
plt.title("Sampling Frequency 50 Hz (Interpolated)")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid()

plt.subplot(236)
plt.plot(t_resampled, cosine3)
plt.title("Sampling Frequency 80 Hz (Interpolated)")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid()

plt.tight_layout()
plt.show()
```
